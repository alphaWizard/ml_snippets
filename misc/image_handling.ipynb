```python
# Load library
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load image as grayscale
image = cv2.imread("arpit.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
# Show image
plt.imshow(image, cmap="gray"), plt.axis("off")
plt.show()
```

```python
# In the matrix, each element contains three values corresponding to blue, green, red values (BGR):
# Load image in color
image_bgr = cv2.imread("arpit.jpg", cv2.IMREAD_COLOR)

# Show 1st pixel(GBR value)
image_bgr[0][0]
```

```python
# Convert to RGB
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Show image
plt.imshow(image_rgb), plt.axis("off")
plt.show()
```

```python
#saving images
# Load libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load image as grayscale
image = cv2.imread("arpit.jpg", cv2.IMREAD_GRAYSCALE)

# Save image
cv2.imwrite("arpit_gray.jpg", image)
```
